```python

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
plt.rcParams["figure.dpi"] = 120



# ---------- 1) Auto-detect files (returns + stock info) ----------
def try_read_table(path):
    ext = os.path.splitext(path)[1].lower()
    try:
        if ext in [".csv"]:
            return pd.read_csv(path)
        elif ext in [".xlsx", ".xls"]:
            return pd.read_excel(path)
        else:
            return None
    except Exception:
        return None

def infer_files(base_path):
    candidates = []
    for fn in os.listdir(base_path):
        if fn.lower().endswith((".csv", ".xlsx", ".xls")):
            full = os.path.join(base_path, fn)
            df = try_read_table(full)
            if df is None:
                continue
            candidates.append((full, df))

    if not candidates:
        raise FileNotFoundError("No CSV/XLSX files found in BASE_PATH")

    # Heuristic:
    # - returns file: many rows (~>200) and many numeric columns (>=10), maybe 30 stocks
    # - info file: fewer rows (~<=100) and text columns like sector/name
    best_returns = None
    best_info = None

    for path, df in candidates:
        # count numeric columns
        num_cols = df.select_dtypes(include=[np.number]).shape[1]
        rows = df.shape[0]
        cols = df.shape[1]

        score_returns = 0
        score_info = 0

        # returns-like
        if rows >= 200: score_returns += 2
        if num_cols >= 10: score_returns += 2
        if cols >= 10: score_returns += 1

        # info-like
        if rows <= 200: score_info += 1
        if cols <= 10: score_info += 1
        # contains typical words
        joined = " ".join([c.lower() for c in df.columns.astype(str)])
        if any(k in joined for k in ["sector", "industry", "name"]): score_info += 3

        if best_returns is None or score_returns > best_returns[0]:
            best_returns = (score_returns, path, df)

        if best_info is None or score_info > best_info[0]:
            best_info = (score_info, path, df)

    returns_path, returns_df = best_returns[1], best_returns[2]
    info_path, info_df = best_info[1], best_info[2]

    # If it picked the same file for both, just set info_df to None (optional)
    if returns_path == info_path:
        info_df = None
        info_path = None

    return returns_path, returns_df, info_path, info_df

RETURNS_PATH, ret_raw, INFO_PATH, info_raw = infer_files(".")
print("\n✅ Detected RETURNS file:", RETURNS_PATH)
if INFO_PATH:
    print("✅ Detected INFO file   :", INFO_PATH)
else:
    print("ℹ️ No separate stock-info file detected (that's okay).")

# ---------- 2) Clean returns table into Date index + tickers columns ----------
ret = ret_raw.copy()

# Try to detect date column
date_col = None
for c in ret.columns:
    if "date" in str(c).lower():
        date_col = c
        break

# If no explicit date column, sometimes first column is date-like
if date_col is None:
    first = ret.columns[0]
    # attempt parse
    try:
        parsed = pd.to_datetime(ret[first], errors="coerce")
        if parsed.notna().mean() > 0.8:
            date_col = first
    except Exception:
        date_col = None

if date_col is not None:
    ret[date_col] = pd.to_datetime(ret[date_col], errors="coerce")
    ret = ret.dropna(subset=[date_col]).sort_values(date_col).set_index(date_col)

# Keep only numeric columns = return series
ret = ret.select_dtypes(include=[np.number]).copy()

# Drop completely empty columns/rows
ret = ret.dropna(axis=1, how="all").dropna(axis=0, how="all")
ret = ret.dropna()  # simplest: remove rows with any missing values

print("\nReturns shape after cleaning:", ret.shape)
print("Sample columns:", ret.columns[:10].tolist())
display(ret.head())

# ---------- 3) FORCE returns to DECIMALS (BULLETPROOF) ----------
def force_decimal_returns(df):
    """
    Convert percent-point returns (e.g., 12.8 = 12.8%) to decimals (0.128).
    Includes hard assertions to catch conversion failures.
    """
    df = df.copy()
    df = df.select_dtypes(include=[np.number])

    max_abs = df.abs().max().max()
    print("\n=== RETURN CONVERSION CHECK ===")
    print("Before conversion: max |daily return| =", max_abs)

    # If max abs > 1, it's definitely percent-points (e.g., 12.8 meaning 12.8%)
    if max_abs > 1:
        df = df / 100.0
        print("✅ Converted by /100")
    else:
        print("✅ Already in decimal format")

    max_abs2 = df.abs().max().max()
    min_val = df.min().min()
    print("After conversion:  max |daily return| =", max_abs2)
    print("After conversion:  min daily return  =", min_val)

    # Hard sanity checks - STOP execution if these fail
    if max_abs2 >= 1:
        raise ValueError(f"❌ FAILED: Still seeing |returns| >= 1 ({max_abs2}). Data is STILL in percent-points!")
    
    if min_val <= -1:
        raise ValueError(f"❌ FAILED: Found daily return <= -1 ({min_val}). This means -100% loss, which is impossible for daily returns!")

    print("✅ VALIDATION PASSED: Data is in proper decimal format")
    print("   Daily returns range: [{:.6f}, {:.6f}]".format(min_val, max_abs2))
    return df

# Apply conversion and validation
ret = force_decimal_returns(ret)

# Display converted data
print("\nPost-conversion statistics (first 5 columns):")
display(ret.describe().T[["min","max","mean","std"]].head())


# ---------- 4) Select 10 stocks ----------
# IMPORTANT: This step and ALL following steps depend on properly converted returns (decimals, not %)
# If you suspect double-conversion, restart kernel and re-run ALL cells from the top
# Option A: manually set your 10 tickers here (must match ret.columns exactly)
# selected_tickers = ["HINDUNILVR","SBIN",...]
selected_tickers = None

def greedy_low_corr_selection(returns_df, n_select=10):
    corr = returns_df.corr().fillna(0)
    tickers = corr.columns.tolist()

    # start with lowest average |corr|
    avg_abs = corr.abs().mean(axis=1)
    chosen = [avg_abs.idxmin()]

    while len(chosen) < n_select:
        remaining = [t for t in tickers if t not in chosen]
        best_t, best_score = None, None
        for t in remaining:
            score = corr.loc[t, chosen].abs().mean()
            if best_score is None or score < best_score:
                best_score = score
                best_t = t
        chosen.append(best_t)
    return chosen

if selected_tickers is None:
    selected_tickers = greedy_low_corr_selection(ret, n_select=10)

print("\n✅ Selected 10 tickers:", selected_tickers)

# CRITICAL: Create fresh returns_10 from properly converted 'ret'
# DO NOT use any old cached variables
returns_10 = ret[selected_tickers].copy()

# Validate returns_10 is in decimals
print("\n=== VALIDATION: returns_10 ===")
print("Shape:", returns_10.shape)
max_abs_return = returns_10.abs().max().max()
min_return = returns_10.min().min()
print("Range: [{:.6f}, {:.6f}]".format(min_return, max_abs_return))

# Last-chance sanity check before proceeding
if (returns_10 <= -1).any().any():
    raise ValueError("❌ FATAL: returns_10 contains values <= -1. Data is NOT in decimal format!")
if max_abs_return >= 1:
    raise ValueError(f"❌ FATAL: returns_10 has |return| >= 1 ({max_abs_return}). Data is still in percent-points!")

print("✅ returns_10 validation passed - safe to proceed")

# ---------- 5) Correlation Matrix Heatmap (Deliverable) ----------
corr10 = returns_10.corr()

plt.figure(figsize=(9,7))
plt.imshow(corr10.values, aspect="auto", cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar(label="Correlation")
plt.xticks(range(10), selected_tickers, rotation=90)
plt.yticks(range(10), selected_tickers)
plt.title("Correlation Matrix Heatmap (10 Selected Assets)")
plt.tight_layout()
plt.show()

avg_abs_pair_corr = corr10.where(~np.eye(10, dtype=bool)).stack().abs().mean()
print("Avg |pairwise correlation|:", float(avg_abs_pair_corr))

# ---------- 6) Train/Test split ----------
TRADING_DAYS = 252
RF_ANNUAL = 0.065

split_n = TRADING_DAYS  # last ~1 year test

# CRITICAL: Create fresh train/test from properly converted returns_10
# These MUST be created from the decimal-format data
train = returns_10.iloc[:-split_n].copy()
test  = returns_10.iloc[-split_n:].copy()

print("\n=== VALIDATION: Train/Test Split ===")
print("Train rows:", len(train), " | Test rows:", len(test))
print("Train range: [{:.6f}, {:.6f}]".format(train.min().min(), train.max().max()))
print("Test range:  [{:.6f}, {:.6f}]".format(test.min().min(), test.max().max()))

# Final check before computing portfolio parameters
if train.abs().max().max() >= 1 or test.abs().max().max() >= 1:
    raise ValueError("❌ FATAL: train or test data still in percent-points!")
print("✅ Train/test validation passed")

# ---------- 7) Estimate mu and cov from TRAIN (annualized) ----------
# CRITICAL: These parameters drive all portfolio optimization
# They MUST be calculated from properly converted decimal returns (not %)
mu = train.mean().values * TRADING_DAYS              # (10,)
cov = train.cov().values * TRADING_DAYS              # (10,10)
cov = cov + 1e-10 * np.eye(cov.shape[0])            # stabilizer

print("Expected annual returns (mu):")
print("  Min: {:.4f}, Max: {:.4f}, Mean: {:.4f}".format(mu.min(), mu.max(), mu.mean()))
print("Annual volatility (std):")
cov_diag_sqrt = np.sqrt(np.diag(cov))
print("  Min: {:.4f}, Max: {:.4f}, Mean: {:.4f}".format(cov_diag_sqrt.min(), cov_diag_sqrt.max(), cov_diag_sqrt.mean()))

def perf_from_weights(w, mu_annual, cov_annual, rf_annual=RF_ANNUAL):
    r = float(w @ mu_annual)
    v = float(np.sqrt(w @ cov_annual @ w))
    s = (r - rf_annual) / v if v > 0 else np.nan
    return r, v, s

# ---------- 8) Monte Carlo Efficient Frontier (>=10k) ----------
def monte_carlo_frontier(mu_annual, cov_annual, rf_annual, n_portfolios=20000, seed=42):
    rng = np.random.default_rng(seed)
    n = len(mu_annual)

    W = np.zeros((n_portfolios, n))
    rets = np.zeros(n_portfolios)
    vols = np.zeros(n_portfolios)
    sharpes = np.zeros(n_portfolios)

    for i in range(n_portfolios):
        w = rng.random(n)
        w = w / w.sum()  # long-only + sum=1
        W[i] = w

        r = float(w @ mu_annual)
        v = float(np.sqrt(w @ cov_annual @ w))
        s = (r - rf_annual) / v if v > 0 else np.nan

        rets[i], vols[i], sharpes[i] = r, v, s

    idx_gmv = int(np.argmin(vols))
    idx_ms  = int(np.nanargmax(sharpes))
    return W, rets, vols, sharpes, idx_gmv, idx_ms

N_PORTFOLIOS = 20000  # >= 10,000 required
W, rets_mc, vols_mc, sharpes_mc, idx_gmv_mc, idx_ms_mc = monte_carlo_frontier(mu, cov, RF_ANNUAL, N_PORTFOLIOS)

w_gmv_mc = W[idx_gmv_mc]
w_ms_mc  = W[idx_ms_mc]

print("\n--- Monte Carlo Picks (from TRAIN estimates) ---")
print("GMV_MC  (AnnRet, AnnVol, Sharpe):", perf_from_weights(w_gmv_mc, mu, cov))
print("MS_MC   (AnnRet, AnnVol, Sharpe):", perf_from_weights(w_ms_mc,  mu, cov))

# ---------- 9) Plot Efficient Frontier (Deliverable) ----------
plt.figure(figsize=(10,6))
sc = plt.scatter(vols_mc, rets_mc, c=sharpes_mc, s=8)
plt.scatter(vols_mc[idx_gmv_mc], rets_mc[idx_gmv_mc], marker="*", s=280, label="GMV (MC)")
plt.scatter(vols_mc[idx_ms_mc],  rets_mc[idx_ms_mc],  marker="*", s=280, label="Max Sharpe (MC)")
plt.xlabel("Annualized Volatility")
plt.ylabel("Annualized Return")
plt.title(f"Efficient Frontier (Monte Carlo, {N_PORTFOLIOS} portfolios)")
plt.colorbar(sc, label="Sharpe")
plt.legend()
plt.tight_layout()
plt.show()

# ---------- 10) Constrained Optimization (Exact GMV + Exact Max Sharpe) ----------
try:
    from scipy.optimize import minimize
except ImportError:
    import subprocess
    import sys
    print("Installing scipy...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "scipy"])
    from scipy.optimize import minimize

n = len(selected_tickers)
bounds = [(0.0, 1.0)] * n
cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0},)
w0 = np.ones(n) / n

def obj_gmv(w, cov_annual):
    return float(w @ cov_annual @ w)

res_gmv = minimize(obj_gmv, w0, args=(cov,), method="SLSQP", bounds=bounds, constraints=cons)
w_gmv_opt = res_gmv.x
w_gmv_opt = np.clip(w_gmv_opt, 0, 1)
w_gmv_opt = w_gmv_opt / w_gmv_opt.sum()

def obj_neg_sharpe(w, mu_annual, cov_annual, rf_annual):
    r = float(w @ mu_annual)
    v = float(np.sqrt(w @ cov_annual @ w))
    s = (r - rf_annual) / v if v > 0 else -1e9
    return -s

res_ms = minimize(obj_neg_sharpe, w0, args=(mu, cov, RF_ANNUAL), method="SLSQP", bounds=bounds, constraints=cons)
w_ms_opt = res_ms.x
w_ms_opt = np.clip(w_ms_opt, 0, 1)
w_ms_opt = w_ms_opt / w_ms_opt.sum()

print("\n--- Optimizer Status ---")
print("GMV_OPT success:", res_gmv.success, "|", res_gmv.message)
print("MS_OPT  success:", res_ms.success,  "|", res_ms.message)

print("\n--- Optimized Portfolios (from TRAIN estimates) ---")
print("GMV_OPT (AnnRet, AnnVol, Sharpe):", perf_from_weights(w_gmv_opt, mu, cov))
print("MS_OPT  (AnnRet, AnnVol, Sharpe):", perf_from_weights(w_ms_opt,  mu, cov))

# ---------- 11) Weight Attribution Tables (Deliverable) ----------
wt = pd.DataFrame({
    "Stock": selected_tickers,
    "GMV_MC": w_gmv_mc,
    "MaxSharpe_MC": w_ms_mc,
    "GMV_OPT": w_gmv_opt,
    "MaxSharpe_OPT": w_ms_opt
})
wt["Diff_OPT (MS - GMV)"] = wt["MaxSharpe_OPT"] - wt["GMV_OPT"]
wt = wt.sort_values("MaxSharpe_OPT", ascending=False)

print("\n✅ Weight Attribution Table (sorted by MaxSharpe_OPT):")
display(wt)

print("Sums:",
      "GMV_MC=", wt["GMV_MC"].sum(),
      "MS_MC=", wt["MaxSharpe_MC"].sum(),
      "GMV_OPT=", wt["GMV_OPT"].sum(),
      "MS_OPT=", wt["MaxSharpe_OPT"].sum())

# ---------- 12) Overlay optimized points on MC frontier (nice for report) ----------
r_gmv_opt, v_gmv_opt, _ = perf_from_weights(w_gmv_opt, mu, cov)
r_ms_opt,  v_ms_opt,  _ = perf_from_weights(w_ms_opt,  mu, cov)

plt.figure(figsize=(10,6))
sc = plt.scatter(vols_mc, rets_mc, c=sharpes_mc, s=8)
plt.scatter(vols_mc[idx_gmv_mc], rets_mc[idx_gmv_mc], marker="*", s=220, label="GMV (MC)")
plt.scatter(vols_mc[idx_ms_mc],  rets_mc[idx_ms_mc],  marker="*", s=220, label="Max Sharpe (MC)")
plt.scatter(v_gmv_opt, r_gmv_opt, marker="X", s=260, label="GMV (OPT)")
plt.scatter(v_ms_opt,  r_ms_opt,  marker="X", s=260, label="Max Sharpe (OPT)")
plt.xlabel("Annualized Volatility")
plt.ylabel("Annualized Return")
plt.title("Efficient Frontier (MC) + Optimized GMV/MaxSharpe")
plt.colorbar(sc, label="Sharpe")
plt.legend()
plt.tight_layout()
plt.show()

# ---------- 13) Final Evaluation Metrics (Train + Test) ----------
def compute_metrics(daily_returns, rf_annual=RF_ANNUAL, trading_days=TRADING_DAYS):
    dr = pd.Series(daily_returns).dropna()

    ann_ret = float(dr.mean() * trading_days)
    ann_vol = float(dr.std(ddof=1) * np.sqrt(trading_days))
    sharpe = (ann_ret - rf_annual) / ann_vol if ann_vol > 0 else np.nan

    downside = dr[dr < 0]
    downside_dev = float(downside.std(ddof=1) * np.sqrt(trading_days)) if len(downside) > 1 else np.nan
    sortino = (ann_ret - rf_annual) / downside_dev if downside_dev and downside_dev > 0 else np.nan

    equity = (1 + dr).cumprod()
    peak = equity.cummax()
    max_dd = float(((equity / peak) - 1).min())  # negative

    calmar = ann_ret / abs(max_dd) if max_dd < 0 else np.nan

    return ann_ret, ann_vol, sharpe, sortino, max_dd, calmar

def portfolio_daily_returns(df_returns, w):
    return df_returns.values @ np.array(w)

# Baseline + all methods
w_equal = np.ones(n) / n
portfolios = {
    "EqualWeight": w_equal,
    "GMV_MC": w_gmv_mc,
    "MaxSharpe_MC": w_ms_mc,
    "GMV_OPT": w_gmv_opt,
    "MaxSharpe_OPT": w_ms_opt
}

rows = []
for name, w in portfolios.items():
    tr = portfolio_daily_returns(train, w)
    te = portfolio_daily_returns(test,  w)
    rows.append([name, "TRAIN", *compute_metrics(tr)])
    rows.append([name, "TEST",  *compute_metrics(te)])

metrics_df = pd.DataFrame(rows, columns=["Portfolio","Split","AnnReturn","AnnVol","Sharpe","Sortino","MaxDrawdown","Calmar"])

# Pretty view in % for returns/vol/dd
pretty = metrics_df.copy()
pretty["AnnReturn"] *= 100
pretty["AnnVol"] *= 100
pretty["MaxDrawdown"] *= 100

print("\n✅ Final Metrics (Return/Vol/MaxDD shown in %):")
display(pretty.sort_values(["Split","Sharpe"], ascending=[True, False]))

# ---------- 14) Equity Curves on TEST ----------
plt.figure(figsize=(10,6))
for name, w in portfolios.items():
    te = portfolio_daily_returns(test, w)
    eq = (1 + pd.Series(te, index=test.index)).cumprod()
    plt.plot(eq, label=name)

plt.title("Equity Curves on TEST (~last 252 trading days)")
plt.xlabel("Date")
plt.ylabel("Growth of 1 unit")
plt.legend()
plt.tight_layout()
plt.show()

print("\nDONE ✅")
```

## 🎯 Alternative: HRP-Based Stock Selection

**Hierarchical Risk Parity (HRP) Clustering Approach:**

Instead of greedy low-correlation selection, we can use **hierarchical clustering** on the correlation matrix:

1. **Compute correlation distance**: $d_{ij} = \sqrt{\frac{1-\rho_{ij}}{2}}$
2. **Hierarchical clustering** to identify correlation groups
3. **Cut into 10 clusters**
4. **Select 1 representative per cluster**:
   - **Medoid**: Most central stock in cluster (pure diversification)
   - **Best Sharpe**: Highest Sharpe within cluster (performance-aware)

**Advantages:**
- ✅ Data-driven correlation analysis (required by assignment)
- ✅ Natural sector diversification without manual rules
- ✅ Avoids picking highly correlated stocks
- ✅ Based on established portfolio theory (HRP by Lopez de Prado)

**For Report:**
> *"We apply hierarchical clustering on correlation distances to partition assets into 10 groups, selecting the most representative stock (medoid) from each cluster. This ensures diversification across correlation structures while being purely data-driven."*

```python
# ============================================================
# HRP-BASED STOCK SELECTION (Hierarchical Clustering)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
from scipy.spatial.distance import squareform

print("="*70)
print("HRP-BASED STOCK SELECTION VIA HIERARCHICAL CLUSTERING")
print("="*70)

def hrp_select_10_stocks(
    returns_30,
    n_select=10,
    linkage_method="single",   # "single" is common in HRP; "average" also ok
    rep_method="medoid",       # "medoid" or "best_sharpe"
    stock_info=None,
    ticker_col_candidates=("Ticker", "Stock", "Symbol", "Name"),
    sector_col_candidates=("Sector", "Industry", "Category"),
    plot_dendrogram=True,
):
    """
    Select n_select stocks using HRP-style hierarchical clustering.
    
    Parameters:
    - returns_30: DataFrame of daily returns for all stocks
    - n_select: Number of stocks to select (default 10)
    - linkage_method: Clustering linkage ('single', 'average', 'complete')
    - rep_method: How to choose representative ('medoid' or 'best_sharpe')
    - stock_info: Optional DataFrame with Ticker and Sector columns
    """
    
    R = returns_30.dropna().copy()
    tickers = R.columns.tolist()
    n = len(tickers)
    
    print(f"\nStarting with {n} stocks")
    print(f"Selection method: {rep_method}")
    print(f"Linkage method: {linkage_method}")
    
    assert n >= n_select, f"Need at least {n_select} columns in returns_30."

    # 1) Correlation matrix
    corr = R.corr().fillna(0)
    print(f"\nAverage correlation: {corr.values[np.triu_indices_from(corr.values, k=1)].mean():.4f}")

    # 2) Correlation-distance: d_ij = sqrt((1 - rho_ij) / 2)
    dist = np.sqrt((1 - corr).clip(-1, 1) / 2)
    dist_condensed = squareform(dist.values, checks=False)

    # 3) Hierarchical clustering
    Z = linkage(dist_condensed, method=linkage_method)

    if plot_dendrogram:
        plt.figure(figsize=(14, 5))
        dendrogram(Z, labels=tickers, leaf_rotation=90, leaf_font_size=10)
        plt.title(f"HRP-style Dendrogram ({linkage_method} linkage)\nCorrelation Distance", fontsize=12)
        plt.xlabel("Stocks")
        plt.ylabel("Distance")
        plt.tight_layout()
        plt.show()

    # 4) Cut into n_select clusters
    labels = fcluster(Z, t=n_select, criterion="maxclust")
    cluster_table = pd.DataFrame({"Ticker": tickers, "Cluster": labels})
    cluster_table["Cluster"] = cluster_table["Cluster"].astype(int)

    # Optional: merge sector info
    if stock_info is not None:
        info = stock_info.copy()
        tcol = next((c for c in ticker_col_candidates if c in info.columns), None)
        scol = next((c for c in sector_col_candidates if c in info.columns), None)
        if tcol and scol:
            info = info[[tcol, scol]].rename(columns={tcol: "Ticker", scol: "Sector"})
            cluster_table = cluster_table.merge(info, on="Ticker", how="left")
        else:
            cluster_table["Sector"] = np.nan
    else:
        cluster_table["Sector"] = np.nan

    # Precompute per-stock stats for "best_sharpe" method
    mu_d = R.mean()
    vol_d = R.std(ddof=1)
    mu_a = mu_d * TRADING_DAYS
    vol_a = vol_d * np.sqrt(TRADING_DAYS)
    sharpe = (mu_a - RF_ANNUAL) / vol_a.replace(0, np.nan)

    stats = pd.DataFrame({
        "Ticker": tickers,
        "AnnRet": mu_a.values,
        "AnnVol": vol_a.values,
        "Sharpe": sharpe.values
    }).set_index("Ticker")

    # Build cluster dictionary
    clusters_dict = {}
    for c in sorted(cluster_table["Cluster"].unique()):
        members = cluster_table.loc[cluster_table["Cluster"] == c, "Ticker"].tolist()
        clusters_dict[c] = members

    print(f"\nFound {len(clusters_dict)} clusters")
    for c, members in clusters_dict.items():
        print(f"  Cluster {c}: {len(members)} stocks - {', '.join(members[:5])}{' ...' if len(members) > 5 else ''}")

    # 5) Choose 1 representative per cluster
    chosen = []
    chosen_sectors = set()

    for c, members in clusters_dict.items():
        if len(members) == 1:
            best = members[0]
        else:
            if rep_method == "best_sharpe":
                best = stats.loc[members, "Sharpe"].idxmax()
            else:  # medoid
                sub = dist.loc[members, members]
                avg_dist = sub.mean(axis=1)
                best = avg_dist.idxmin()

        # Tie-break with sector diversity if possible
        if stock_info is not None and "Sector" in cluster_table.columns:
            sector_map = cluster_table.set_index("Ticker")["Sector"].to_dict()
            if sector_map.get(best) in chosen_sectors:
                alternatives = [m for m in members if sector_map.get(m) not in chosen_sectors]
                if alternatives:
                    if rep_method == "best_sharpe":
                        best = stats.loc[alternatives, "Sharpe"].idxmax()
                    else:
                        sub_alt = dist.loc[alternatives, alternatives]
                        avg_dist_alt = sub_alt.mean(axis=1)
                        best = avg_dist_alt.idxmin()

            if sector_map.get(best) is not None:
                chosen_sectors.add(sector_map.get(best))

        chosen.append(best)

    # Fill remaining if needed (shouldn't happen with maxclust, but safety)
    chosen = list(dict.fromkeys(chosen))
    if len(chosen) < n_select:
        remaining = [t for t in tickers if t not in chosen]
        while len(chosen) < n_select and remaining:
            scores = [(corr.loc[t, chosen].abs().mean(), t) for t in remaining]
            scores.sort()
            chosen.append(scores[0][1])
            remaining.remove(scores[0][1])

    # Trim if too many (rare)
    if len(chosen) > n_select:
        subcorr = corr.loc[chosen, chosen].abs()
        avg_abs = subcorr.mean(axis=1).sort_values()
        chosen = avg_abs.index[:n_select].tolist()

    # Final output table
    out = cluster_table.copy()
    out["Selected"] = out["Ticker"].isin(chosen)
    out = out.sort_values(["Selected", "Cluster", "Ticker"], ascending=[False, True, True])

    return chosen, out, corr, clusters_dict

# ============================================================
# RUN HRP SELECTION
# ============================================================

# Use training period for selection (avoid look-ahead bias)
# Split data to get train period
split_n = 252
train_all = ret.iloc[:-split_n].copy()

print(f"\nUsing train period: {len(train_all)} days")
print(f"Number of stocks available: {len(train_all.columns)}")

# Run HRP selection
selected_tickers_hrp, cluster_table_hrp, corr_all, clusters_hrp = hrp_select_10_stocks(
    train_all,
    n_select=10,
    linkage_method="single",    # HRP typically uses 'single'
    rep_method="best_sharpe",        # Pure diversification (or use "best_sharpe")
    stock_info=info_raw if INFO_PATH else None,
    plot_dendrogram=True
)

print("\n" + "="*70)
print("SELECTED 10 STOCKS (HRP METHOD)")
print("="*70)
print(selected_tickers_hrp)

print("\n✅ Cluster Assignment Table:")
display(cluster_table_hrp)

# Show sector distribution if available
if "Sector" in cluster_table_hrp.columns and cluster_table_hrp["Sector"].notna().any():
    print("\n📊 Sector Distribution Among Selected:")
    sector_dist = cluster_table_hrp[cluster_table_hrp["Selected"]]["Sector"].value_counts()
    display(sector_dist)
    print(f"\nUnique sectors: {sector_dist.shape[0]}")

# Correlation heatmap for selected 10
print("\n" + "="*70)
print("CORRELATION HEATMAP (HRP-SELECTED 10 STOCKS)")
print("="*70)

corr10_hrp = train_all[selected_tickers_hrp].corr()

plt.figure(figsize=(10, 8))
im = plt.imshow(corr10_hrp.values, aspect="auto", cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar(im, label="Correlation")
plt.xticks(range(10), selected_tickers_hrp, rotation=90)
plt.yticks(range(10), selected_tickers_hrp)
plt.title("Correlation Heatmap (HRP-Selected 10 Stocks)\nMedian Absolute Correlation", fontsize=12)

# Add correlation values as text
for i in range(10):
    for j in range(10):
        text = plt.text(j, i, f'{corr10_hrp.values[i, j]:.2f}',
                       ha="center", va="center", color="black", fontsize=8)

plt.tight_layout()
plt.show()

avg_abs_corr_hrp = corr10_hrp.where(~np.eye(10, dtype=bool)).stack().abs().mean()
print(f"\n📈 Average |pairwise correlation|: {avg_abs_corr_hrp:.4f}")
print(f"📈 Max correlation: {corr10_hrp.where(~np.eye(10, dtype=bool)).max().max():.4f}")
print(f"📈 Min correlation: {corr10_hrp.where(~np.eye(10, dtype=bool)).min().min():.4f}")

print("\n✅ HRP-based selection complete!")
print("\n💡 You can now use 'selected_tickers_hrp' instead of 'selected_tickers' for your analysis")
print("   Or compare both methods by running the portfolio optimization twice")
```

### Why HRP?

- **Theoretically rigorous**: Based on hierarchical clustering literature

## 🔧 Weight Constraints for Diversification

Even with HRP-based stock selection, **mean-variance optimization** can produce concentrated portfolios due to:
- **Estimation error** in expected returns (μ)
- **Optimizer exploiting noise** in the training data
- **Extreme weights** that don't generalize well out-of-sample

### Solution: Maximum Weight Constraints

Add a **maximum weight constraint** (e.g., ≤25% per stock) to:
- ✅ Force diversification across selected stocks
- ✅ Reduce sensitivity to parameter estimation errors
- ✅ Improve out-of-sample robustness
- ✅ Create more practical, implementable portfolios

**Justification for Report:**
*"Unconstrained mean-variance optimization can yield highly concentrated portfolios due to estimation noise in expected returns. We impose a practical concentration limit (wᵢ ≤ 25%) to improve diversification and out-of-sample stability while maintaining the long-only and budget constraints."*

This is a **standard practice** in institutional portfolio management to prevent over-concentration risk.

## 🔄 Rolling Rebalancing Backtest

The previous analysis used a single train/test split. However, real portfolio management requires **periodic rebalancing** with updated estimates. 

This section implements a **realistic monthly rebalancing backtest**:
- **Rebalances every ~21 trading days** (monthly)
- **Rolling estimation window** (252 days lookback)
- **Out-of-sample evaluation** on last 252 trading days
- **Tests all strategies** with realistic implementation

This addresses the key question: *"Do these portfolios maintain performance when parameters are re-estimated over time?"*

```python
# ============================================================
# ROLLING REBALANCING BACKTEST (MONTHLY)
# - Uses rolling lookback window to estimate mu/cov
# - Rebalances every ~21 trading days
# - Evaluates on last TEST_DAYS (out-of-sample)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# -------------------- PARAMETERS --------------------
TRADING_DAYS   = 252
RF_ANNUAL      = 0.065

TEST_DAYS      = 252     # evaluate last ~1 year
LOOKBACK_DAYS  = 252     # estimation window (try 504 for more stability if you want)
REBAL_EVERY    = 21      # monthly-ish rebalance

N_MC           = 20000   # Monte Carlo portfolios each rebalance (>=10k is fine)
MC_SEED        = 42

MAX_W_CAP      = 0.25    # cap for capped MaxSharpe optimization (try 0.20 too)

# -------------------- SANITY CHECK --------------------
print("="*70)
print("ROLLING REBALANCING BACKTEST SETUP")
print("="*70)
assert returns_10.abs().max().max() < 1, "returns_10 still looks like percent points. Divide by 100 first."
returns_10_clean = returns_10.dropna().copy()
tickers = returns_10_clean.columns.tolist()
n = len(tickers)

print("\nAssets:", tickers)
print("Data rows:", len(returns_10_clean))
print("Min daily return:", float(returns_10_clean.min().min()))
print("Max abs daily return:", float(returns_10_clean.abs().max().max()))
print("\nBacktest config:")
print(f"  - Test period: {TEST_DAYS} days")
print(f"  - Lookback window: {LOOKBACK_DAYS} days")
print(f"  - Rebalance frequency: {REBAL_EVERY} days (~monthly)")
print(f"  - Monte Carlo portfolios per rebalance: {N_MC}")
print(f"  - Weight cap: {MAX_W_CAP:.0%}")

# -------------------- HELPERS --------------------
def estimate_mu_cov(window_returns):
    """Estimate annualized mu and cov from a window of daily returns."""
    mu = window_returns.mean().values * TRADING_DAYS
    cov = window_returns.cov().values * TRADING_DAYS
    cov = cov + 1e-10 * np.eye(cov.shape[0])  # stabilizer
    return mu, cov

def perf_from_weights_roll(w, mu, cov, rf=RF_ANNUAL):
    r = float(w @ mu)
    v = float(np.sqrt(w @ cov @ w))
    s = (r - rf) / v if v > 0 else np.nan
    return r, v, s

def w_equal_roll(n):
    return np.ones(n) / n

def w_gmv_opt_roll(cov, max_w=1.0):
    bounds = [(0.0, max_w)] * n
    cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0},)
    w0 = np.ones(n) / n

    def obj(w): return float(w @ cov @ w)

    res = minimize(obj, w0, method="SLSQP", bounds=bounds, constraints=cons)
    w = res.x
    w = np.clip(w, 0, max_w)
    w = w / w.sum()
    return w

def w_maxsharpe_opt_roll(mu, cov, rf=RF_ANNUAL, max_w=1.0):
    bounds = [(0.0, max_w)] * n
    cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0},)
    w0 = np.ones(n) / n

    def obj_neg_sharpe(w):
        r = float(w @ mu)
        v = float(np.sqrt(w @ cov @ w))
        s = (r - rf) / v if v > 0 else -1e9
        return -s

    res = minimize(obj_neg_sharpe, w0, method="SLSQP", bounds=bounds, constraints=cons)
    w = res.x
    w = np.clip(w, 0, max_w)
    w = w / w.sum()
    return w

def w_mc_best_roll(mu, cov, rf=RF_ANNUAL, n_portfolios=N_MC, seed=MC_SEED, mode="max_sharpe"):
    rng = np.random.default_rng(seed)
    best_w = None
    best_val = None

    for _ in range(n_portfolios):
        w = rng.random(n)
        w = w / w.sum()

        r = float(w @ mu)
        v = float(np.sqrt(w @ cov @ w))

        if mode == "gmv":
            val = v
            better = (best_val is None) or (val < best_val)
        else:  # max_sharpe
            s = (r - rf) / v if v > 0 else -np.inf
            val = s
            better = (best_val is None) or (val > best_val)

        if better:
            best_val = val
            best_w = w

    return best_w

def compute_metrics_roll(daily_returns, rf_annual=RF_ANNUAL, trading_days=TRADING_DAYS):
    dr = pd.Series(daily_returns).dropna()

    ann_ret = float(dr.mean() * trading_days)
    ann_vol = float(dr.std(ddof=1) * np.sqrt(trading_days))
    sharpe = (ann_ret - rf_annual) / ann_vol if ann_vol > 0 else np.nan

    downside = dr[dr < 0]
    downside_dev = float(downside.std(ddof=1) * np.sqrt(trading_days)) if len(downside) > 1 else np.nan
    sortino = (ann_ret - rf_annual) / downside_dev if downside_dev and downside_dev > 0 else np.nan

    equity = (1 + dr).cumprod()
    peak = equity.cummax()
    max_dd = float(((equity / peak) - 1).min())  # negative

    calmar = ann_ret / abs(max_dd) if max_dd < 0 else np.nan

    return {
        "AnnReturn": ann_ret,
        "AnnVol": ann_vol,
        "Sharpe": sharpe,
        "Sortino": sortino,
        "MaxDrawdown": max_dd,
        "Calmar": calmar
    }

# -------------------- ROLLING BACKTEST CORE --------------------
def rolling_backtest(returns, weight_func, name="strategy"):
    """
    returns: full returns df (decimal daily)
    weight_func: function(window_returns, mu, cov)->weights
    rebalances every REBAL_EVERY days using last LOOKBACK_DAYS.
    Evaluates only over last TEST_DAYS (out-of-sample).
    """
    T = len(returns)
    start_test = T - TEST_DAYS

    # First rebalance date must have enough lookback and be within test period
    first_rebal = max(start_test, LOOKBACK_DAYS)
    # Align to a rebalance grid (optional)
    rebal_dates = list(range(first_rebal, T, REBAL_EVERY))
    if rebal_dates[-1] != T:
        rebal_dates.append(T)

    port_rets = []

    # iterate each holding period
    for i in range(len(rebal_dates) - 1):
        t_rebal = rebal_dates[i]
        t_next  = rebal_dates[i+1]

        window = returns.iloc[t_rebal-LOOKBACK_DAYS:t_rebal]
        mu, cov = estimate_mu_cov(window)

        w = weight_func(window, mu, cov)

        hold = returns.iloc[t_rebal:t_next]
        pr = hold.values @ w
        port_rets.append(pd.Series(pr, index=hold.index))

    port = pd.concat(port_rets).iloc[:TEST_DAYS]  # ensure exact test length
    port.name = name
    return port

# -------------------- DEFINE STRATEGIES --------------------
# Each weight_func signature: (window_returns, mu, cov) -> w

def wf_equal(window, mu, cov):
    return w_equal_roll(n)

def wf_gmv_opt_old(window, mu, cov):
    return w_gmv_opt_roll(cov, max_w=1.0)

def wf_gmv_opt_cap(window, mu, cov):
    return w_gmv_opt_roll(cov, max_w=MAX_W_CAP)

def wf_ms_opt_old(window, mu, cov):
    return w_maxsharpe_opt_roll(mu, cov, rf=RF_ANNUAL, max_w=1.0)

def wf_ms_opt_cap(window, mu, cov):
    return w_maxsharpe_opt_roll(mu, cov, rf=RF_ANNUAL, max_w=MAX_W_CAP)

def wf_gmv_mc(window, mu, cov):
    # vary seed by time index for diversity
    return w_mc_best_roll(mu, cov, rf=RF_ANNUAL, n_portfolios=N_MC, seed=MC_SEED + len(window), mode="gmv")

def wf_ms_mc(window, mu, cov):
    return w_mc_best_roll(mu, cov, rf=RF_ANNUAL, n_portfolios=N_MC, seed=MC_SEED + len(window), mode="max_sharpe")

strategies = {
    "EqualWeight": wf_equal,
    "GMV_OPT": wf_gmv_opt_old,
    "MaxSharpe_OPT": wf_ms_opt_old,
    f"MaxSharpe_OPT_cap_{int(MAX_W_CAP*100)}%": wf_ms_opt_cap,
    "GMV_MC": wf_gmv_mc,
    "MaxSharpe_MC": wf_ms_mc,
}

# -------------------- RUN BACKTEST --------------------
print("\n" + "="*70)
print("RUNNING ROLLING BACKTEST...")
print("="*70)

results = {}
for name, wf in strategies.items():
    print(f"  Running: {name}...")
    results[name] = rolling_backtest(returns_10_clean, wf, name=name)

# Combine portfolio return series
port_df = pd.concat(results.values(), axis=1)

print("\n✅ Backtest complete! Sample portfolio returns:")
display(port_df.head())

# -------------------- METRICS TABLE (TEST ONLY, rolling) --------------------
print("\n" + "="*70)
print("COMPUTING PERFORMANCE METRICS")
print("="*70)

rows = []
for name in port_df.columns:
    m = compute_metrics_roll(port_df[name])
    rows.append([name, m["AnnReturn"], m["AnnVol"], m["Sharpe"], m["Sortino"], m["MaxDrawdown"], m["Calmar"]])

metrics_roll = pd.DataFrame(rows, columns=["Portfolio","AnnReturn","AnnVol","Sharpe","Sortino","MaxDrawdown","Calmar"])
metrics_roll = metrics_roll.sort_values("Sharpe", ascending=False)

# pretty in %
pretty_roll = metrics_roll.copy()
pretty_roll["AnnReturn"] *= 100
pretty_roll["AnnVol"] *= 100
pretty_roll["MaxDrawdown"] *= 100

print("\n✅ ROLLING TEST METRICS (Return/Vol/MaxDD in %):")
display(pretty_roll)

print("\n📊 KEY OBSERVATIONS:")
print(f"   Best Sharpe: {pretty_roll.iloc[0]['Portfolio']} ({pretty_roll.iloc[0]['Sharpe']:.4f})")
print(f"   Lowest Vol: {pretty_roll.loc[pretty_roll['AnnVol'].idxmin()]['Portfolio']} ({pretty_roll['AnnVol'].min():.2f}%)")
print(f"   Lowest Drawdown: {pretty_roll.loc[pretty_roll['MaxDrawdown'].idxmax()]['Portfolio']} ({pretty_roll['MaxDrawdown'].max():.2f}%)")

# -------------------- EQUITY CURVES (TEST) --------------------
plt.figure(figsize=(12,7))
for name in port_df.columns:
    eq = (1 + port_df[name]).cumprod()
    linewidth = 2.5 if 'cap' in name else 2.0
    linestyle = '-' if 'OPT' in name or 'Equal' in name else '--'
    plt.plot(eq, label=name, linewidth=linewidth, linestyle=linestyle)

plt.title(f"Rolling Backtest Equity Curves\n(TEST={TEST_DAYS}d, Lookback={LOOKBACK_DAYS}d, Rebalance every {REBAL_EVERY}d)", 
          fontsize=13)
plt.xlabel("Date")
plt.ylabel("Growth of $1")
plt.legend(loc='best', fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\n✅ Rolling backtest visualization complete!")
```

## 🔬 Advanced: Shrinkage Estimators

**Problem with Sample Estimates:** 
- Sample mean (μ) is extremely noisy with limited data
- Sample covariance can be poorly conditioned

**Solution: Shrinkage**
1. **Ledoit-Wolf Covariance**: Shrinks sample covariance toward structured target (improves stability)
2. **Mean Shrinkage**: Shrinks individual means toward cross-sectional average (reduces estimation error)

**Formula:**
- μ_shrunk = α × μ_sample + (1-α) × μ_prior
- Where α ∈ [0,1] controls shrinkage intensity

**Expected Impact:**
- Better out-of-sample Sharpe ratios
- More stable weight allocations
- Reduced turnover at rebalancing

*This is a well-established technique in quantitative finance (DeMiguel et al., 2009).*

```python
# ============================================================
# UPGRADE: Ledoit–Wolf Covariance + Mean Shrinkage
# Improves out-of-sample stability by reducing estimation error
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.covariance import LedoitWolf

# -------------------- PARAMETERS --------------------
TRADING_DAYS   = 252
RF_ANNUAL      = 0.065

TEST_DAYS      = 252
LOOKBACK_DAYS  = 252     # try 504 for even more stability
REBAL_EVERY    = 21

N_MC           = 20000
MC_SEED        = 42

MAX_W_CAP      = 0.25    # keep the cap version too

# Mean shrinkage strength (0 = all prior, 1 = all sample mean)
ALPHA_MU       = 0.60    # good starting point; try 0.2–0.6

print("="*70)
print("SHRINKAGE ESTIMATORS: Ledoit-Wolf + Mean Shrinkage")
print("="*70)

# -------------------- SANITY CHECK --------------------
assert returns_10.abs().max().max() < 1, "returns_10 looks like percent points. Divide by 100 first."
returns_10_clean2 = returns_10.dropna().copy()
tickers = returns_10_clean2.columns.tolist()
n = len(tickers)

print(f"\nAssets: {tickers}")
print(f"Rows: {len(returns_10_clean2)} | Max |daily|: {float(returns_10_clean2.abs().max().max()):.6f}")
print(f"\nShrinkage config:")
print(f"  - Mean shrinkage intensity (ALPHA_MU): {ALPHA_MU:.2f}")
print(f"  - Covariance: Ledoit-Wolf automatic shrinkage")

# -------------------- ESTIMATORS --------------------
def estimate_mu_cov_sample(window_returns):
    """Sample mean + sample covariance (annualized)."""
    mu_d = window_returns.mean().values
    cov_d = window_returns.cov().values
    mu = mu_d * TRADING_DAYS
    cov = cov_d * TRADING_DAYS + 1e-10*np.eye(cov_d.shape[0])
    return mu, cov

def estimate_mu_cov_lw_shrunk(window_returns, alpha_mu=ALPHA_MU):
    """
    Ledoit–Wolf covariance shrinkage + mean shrinkage.
    Mean shrinkage prior = cross-sectional mean (same mean for all assets).
    """
    X = window_returns.values  # shape (T, n)

    # ---- mean (daily) ----
    mu_d = X.mean(axis=0)  # sample daily means, shape (n,)
    mu_prior = np.full(n, mu_d.mean())  # cross-sectional mean prior
    mu_d_shrunk = alpha_mu * mu_d + (1 - alpha_mu) * mu_prior

    # ---- covariance (daily) via Ledoit–Wolf ----
    lw = LedoitWolf().fit(X)
    cov_d_lw = lw.covariance_

    # annualize
    mu = mu_d_shrunk * TRADING_DAYS
    cov = cov_d_lw * TRADING_DAYS + 1e-10*np.eye(n)

    return mu, cov

# -------------------- PORTFOLIO HELPERS --------------------
def perf_from_weights_shrink(w, mu, cov, rf=RF_ANNUAL):
    r = float(w @ mu)
    v = float(np.sqrt(w @ cov @ w))
    s = (r - rf) / v if v > 0 else np.nan
    return r, v, s

def w_equal_shrink():
    return np.ones(n) / n

def w_gmv_opt_shrink(cov, max_w=1.0):
    bounds = [(0.0, max_w)] * n
    cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0},)
    w0 = np.ones(n) / n

    def obj(w): return float(w @ cov @ w)

    res = minimize(obj, w0, method="SLSQP", bounds=bounds, constraints=cons)
    w = np.clip(res.x, 0, max_w)
    w = w / w.sum()
    return w

def w_maxsharpe_opt_shrink(mu, cov, rf=RF_ANNUAL, max_w=1.0):
    bounds = [(0.0, max_w)] * n
    cons = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0},)
    w0 = np.ones(n) / n

    def obj_neg_sharpe(w):
        r = float(w @ mu)
        v = float(np.sqrt(w @ cov @ w))
        s = (r - rf) / v if v > 0 else -1e9
        return -s

    res = minimize(obj_neg_sharpe, w0, method="SLSQP", bounds=bounds, constraints=cons)
    w = np.clip(res.x, 0, max_w)
    w = w / w.sum()
    return w

def w_mc_best_shrink(mu, cov, rf=RF_ANNUAL, n_portfolios=N_MC, seed=MC_SEED, mode="max_sharpe"):
    rng = np.random.default_rng(seed)
    best_w, best_val = None, None

    for _ in range(n_portfolios):
        w = rng.random(n)
        w = w / w.sum()

        r = float(w @ mu)
        v = float(np.sqrt(w @ cov @ w))

        if mode == "gmv":
            val = v
            better = (best_val is None) or (val < best_val)
        else:
            s = (r - rf) / v if v > 0 else -np.inf
            val = s
            better = (best_val is None) or (val > best_val)

        if better:
            best_val = val
            best_w = w

    return best_w

# -------------------- METRICS --------------------
def compute_metrics_shrink(daily_returns, rf_annual=RF_ANNUAL, trading_days=TRADING_DAYS):
    dr = pd.Series(daily_returns).dropna()
    ann_ret = float(dr.mean() * trading_days)
    ann_vol = float(dr.std(ddof=1) * np.sqrt(trading_days))
    sharpe = (ann_ret - rf_annual) / ann_vol if ann_vol > 0 else np.nan

    downside = dr[dr < 0]
    downside_dev = float(downside.std(ddof=1) * np.sqrt(trading_days)) if len(downside) > 1 else np.nan
    sortino = (ann_ret - rf_annual) / downside_dev if downside_dev and downside_dev > 0 else np.nan

    equity = (1 + dr).cumprod()
    peak = equity.cummax()
    max_dd = float(((equity / peak) - 1).min())

    calmar = ann_ret / abs(max_dd) if max_dd < 0 else np.nan
    return ann_ret, ann_vol, sharpe, sortino, max_dd, calmar

# -------------------- ROLLING BACKTEST (generic estimator) --------------------
def rolling_backtest_shrink(returns, estimator_fn, weight_rule_fn, name="strategy"):
    """
    estimator_fn(window_returns) -> (mu, cov)
    weight_rule_fn(window_returns, mu, cov, t_rebal) -> weights
    """
    T = len(returns)
    start_test = T - TEST_DAYS
    first_rebal = max(start_test, LOOKBACK_DAYS)

    rebal_idx = list(range(first_rebal, T, REBAL_EVERY))
    if rebal_idx[-1] != T:
        rebal_idx.append(T)

    port_rets = []
    for i in range(len(rebal_idx) - 1):
        t_rebal, t_next = rebal_idx[i], rebal_idx[i+1]
        window = returns.iloc[t_rebal-LOOKBACK_DAYS:t_rebal]

        mu, cov = estimator_fn(window)
        w = weight_rule_fn(window, mu, cov, t_rebal)

        hold = returns.iloc[t_rebal:t_next]
        pr = hold.values @ w
        port_rets.append(pd.Series(pr, index=hold.index))

    port = pd.concat(port_rets).iloc[:TEST_DAYS]
    port.name = name
    return port

# -------------------- STRATEGIES: SAMPLE vs SHRINKAGE --------------------
# weight_rule signature: (window, mu, cov, t_rebal) -> w
def wr_equal_s(window, mu, cov, t): return w_equal_shrink()

# Sample-estimator strategies
def wr_gmv_opt_s(window, mu, cov, t): return w_gmv_opt_shrink(cov, max_w=1.0)
def wr_ms_opt_s(window, mu, cov, t):  return w_maxsharpe_opt_shrink(mu, cov, rf=RF_ANNUAL, max_w=1.0)
def wr_ms_opt_cap_s(window, mu, cov, t): return w_maxsharpe_opt_shrink(mu, cov, rf=RF_ANNUAL, max_w=MAX_W_CAP)
def wr_gmv_mc_s(window, mu, cov, t): return w_mc_best_shrink(mu, cov, rf=RF_ANNUAL, n_portfolios=N_MC, seed=MC_SEED+t, mode="gmv")
def wr_ms_mc_s(window, mu, cov, t):  return w_mc_best_shrink(mu, cov, rf=RF_ANNUAL, n_portfolios=N_MC, seed=MC_SEED+t, mode="max_sharpe")

strategies = [
    # Baseline
    ("EqualWeight", estimate_mu_cov_sample, wr_equal_s),

    # Monte Carlo (sample estimator)
    ("MaxSharpe_MC_sample", estimate_mu_cov_sample, wr_ms_mc_s),
    ("GMV_MC_sample",       estimate_mu_cov_sample, wr_gmv_mc_s),

    # Optimized (sample estimator)
    ("MaxSharpe_OPT_sample",      estimate_mu_cov_sample, wr_ms_opt_s),
    (f"MaxSharpe_OPT_cap{int(MAX_W_CAP*100)}_sample", estimate_mu_cov_sample, wr_ms_opt_cap_s),
    ("GMV_OPT_sample",            estimate_mu_cov_sample, wr_gmv_opt_s),

    # Monte Carlo (LW + mean shrinkage)
    ("MaxSharpe_MC_LWshrink", estimate_mu_cov_lw_shrunk, wr_ms_mc_s),
    ("GMV_MC_LWshrink",       estimate_mu_cov_lw_shrunk, wr_gmv_mc_s),

    # Optimized (LW + mean shrinkage)
    ("MaxSharpe_OPT_LWshrink",      estimate_mu_cov_lw_shrunk, wr_ms_opt_s),
    (f"MaxSharpe_OPT_cap{int(MAX_W_CAP*100)}_LWshrink", estimate_mu_cov_lw_shrunk, wr_ms_opt_cap_s),
    ("GMV_OPT_LWshrink",            estimate_mu_cov_lw_shrunk, wr_gmv_opt_s),
]

# -------------------- RUN --------------------
print("\n" + "="*70)
print("RUNNING BACKTEST WITH SHRINKAGE ESTIMATORS...")
print("="*70)

results_shrink = {}
for name, est, rule in strategies:
    print(f"  Running: {name}...")
    results_shrink[name] = rolling_backtest_shrink(returns_10_clean2, est, rule, name=name)

port_df_shrink = pd.concat(results_shrink.values(), axis=1)

print("\n✅ Sample portfolio returns:")
display(port_df_shrink.head())

# -------------------- METRICS --------------------
print("\n" + "="*70)
print("COMPUTING METRICS...")
print("="*70)

rows_shrink = []
for name in port_df_shrink.columns:
    ann_ret, ann_vol, sharpe, sortino, mdd, calmar = compute_metrics_shrink(port_df_shrink[name])
    rows_shrink.append([name, ann_ret, ann_vol, sharpe, sortino, mdd, calmar])

metrics_shrink = pd.DataFrame(rows_shrink, columns=["Portfolio","AnnReturn","AnnVol","Sharpe","Sortino","MaxDrawdown","Calmar"])
metrics_shrink = metrics_shrink.sort_values("Sharpe", ascending=False)

pretty_shrink = metrics_shrink.copy()
pretty_shrink["AnnReturn"] *= 100
pretty_shrink["AnnVol"] *= 100
pretty_shrink["MaxDrawdown"] *= 100

print("\n✅ ROLLING TEST METRICS WITH SHRINKAGE (Return/Vol/MaxDD in %):")
display(pretty_shrink)

# -------------------- COMPARISON: SAMPLE vs SHRINKAGE --------------------
print("\n📊 SAMPLE vs SHRINKAGE Comparison:")
print("\nSample Estimator Sharpes:")
sample_results = pretty_shrink[pretty_shrink['Portfolio'].str.contains('sample')]
for idx, row in sample_results.iterrows():
    print(f"  {row['Portfolio']:35s}: {row['Sharpe']:.4f}")

print("\nLedoit-Wolf Shrinkage Sharpes:")
shrink_results = pretty_shrink[pretty_shrink['Portfolio'].str.contains('LWshrink')]
for idx, row in shrink_results.iterrows():
    print(f"  {row['Portfolio']:35s}: {row['Sharpe']:.4f}")

# -------------------- PLOT EQUITY CURVES (top portfolios) --------------------
topN = 8
top_names = metrics_shrink["Portfolio"].head(topN).tolist()

plt.figure(figsize=(12,7))
for name in top_names:
    eq = (1 + port_df_shrink[name]).cumprod()
    linewidth = 2.5 if 'LWshrink' in name else 1.8
    linestyle = '-' if 'cap' in name or 'LWshrink' in name else '--'
    plt.plot(eq, label=name, linewidth=linewidth, linestyle=linestyle)

plt.title(f"Top {topN} Strategies: Sample vs Shrinkage Estimators\n(TEST={TEST_DAYS}d, Rebalance={REBAL_EVERY}d, α_μ={ALPHA_MU})", 
         fontsize=12)
plt.xlabel("Date")
plt.ylabel("Growth of $1")
plt.legend(loc='best', fontsize=9)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 TIP: Try ALPHA_MU = 0.2, 0.3, 0.5 and LOOKBACK_DAYS = 504 to optimize further.")
print("✅ Shrinkage backtest complete!")
```

### 📈 Shrinkage Results Analysis

**What to Look For:**

1. **Improved Sharpe Ratios**: Shrinkage versions (LWshrink) should show higher Sharpe than sample versions
2. **Better Stability**: Lower volatility and smaller drawdowns
3. **More Consistent**: Less sensitivity to noisy parameter estimates

**Why Shrinkage Works:**

- **Covariance Shrinkage (Ledoit-Wolf)**: 
  - Sample covariance matrix can have large eigenvalues → extreme allocations
  - Shrinkage "regularizes" → more balanced weights
  
- **Mean Shrinkage (α = 0.3)**:
  - Individual stock means are very noisy with 252 days of data
  - Shrinking toward cross-sectional average reduces overfitting
  - α = 0.3 means 30% sample, 70% prior → conservative

**Tuning Recommendations:**

| Parameter | Range | Effect |
|-----------|-------|--------|
| ALPHA_MU | 0.2-0.6 | Lower = more shrinkage, more stable |
| LOOKBACK_DAYS | 252-504 | Higher = more stable, less adaptive |
| REBAL_EVERY | 21-42 | Lower = more adaptive, higher turnover |

**Academic Support:**

DeMiguel, Garlappi, & Uppal (2009) showed that simple shrinkage often outperforms complex methods due to reduced estimation error.
